```python
def partition_data(X, ratio=[0.8, 0.2]): 

    assert(np.sum(ratio) == 1)

    # shuffle the data indices 
    idxs = np.random.permutation(X.index)

    # the number of samples 
    N = X.shape[0]

    Xs = []
    i = 0  # first index to zero
    for k, r in enumerate(ratio):
        nrows = int(round(N * r))  # number of rows

        if k == len(ratio) -1:
            Xs.append(X.iloc[i:, :])
        else:
            Xs.append(X.iloc[i:i+nrows, :])

        i += nrows

    return Xs
```

# Method

###  1.Pocket Algorithm
To understand pocket algorithm we will have to go through perceptron algorithm first. The perceptron algorithm uses a single perceptron. A perceptron is a linear model which applies a step function on the linear output. To perform classification, it iterates over and over the input data and keep adjusting weights to minimize the errors in the predicted and the actual output values. The weight adjustment for the process looks like below 

$$
\wv^{(k+1)} = \wv^{(k)} + \alpha t_i \xv_i
$$

In the Perceptron algorithm, the algo might overlook a great solution to find better solutions. To overcome this instability, pocket algorithm comes into the picture. The pocket algorithm checks the prediction from new weights against the best prediction so far. The weights are updated in case this new prediction is better and if that is not the case, the existing best prediction weights are maintained. 

<b>Implementation:</b> 

1. Train Method: The train method trains takes 4 parameters which are the training data X, the actual output T, the number of iterations and the learning rate. Following steps are performed in the method: 
    1. We first initialize w(weights) to an array of zeroes. We also declare a w_pocket and then copy the values of w in w_pocket for initialization.
    2. We start a for loop with the number of iterations as passed in the function. Inside the loop we declare a variable converge which is used to check convergence. We also start another for loop which runs equal to the number of records in the input data.
    3. Inside the above loop, we first predict the output with the existing weight. We then compare the predicted y with the actual output. If the predicted value is not equal we recalculate the wrights as w += alpha * T[k] * X[k]. We also mark converged as true. 
    4. In the next step, we call the compare method which compares the results from current weights and the existing best prediction weights. This is where pocket does its magic. 
    5. We check if the converged value is true. If the solution is converged, we break the loop and return the weights in w_pocket. 
    
2. Use Method: The use method takes two variables, the input data X and the weights w. The function returns the dot product of input variables with the weights i.e. predicted Value.

3. Compare Method: The compare method compares the prediction from current weights to the prediction from W_pocket. If the prediction from current weight is better, the w_pocket is updated. If not, the w_pocket remains same.      
    
### 2. Logistic Regression
Logistic Regression is a linear regression model that predicts P(T=k∣x) where k is the class label given given x is the observation.This method does not predict the label directly.The model used for Logistic Regression is as below: 

$$
\kappa = f(\xv ; \wv) = \Xm \wv.
$$

As per gradient descentm we need to update the weights. We do this as below: 


$$
\wv \leftarrow \wv - \alpha \nabla_\wv E(\wv).
$$

We use the softmax function as given below:

$$
g_k(\xv) = P(T=k \mid \xv) = \frac{e^{\kappa_k}}{\sum_{c=1}^K e^{\kappa_c}}
$$

And hence the weight updation becomes:


$$
\wv \leftarrow \wv + \alpha \Xm^\top \Big(  \Tm - g(\Xm)\Big).
$$

<b>Implementation:</b> 
1. Softmax Method: The function takes one input variable, Z matrix which is the dot product of input data X and the weights w.This function applies softmax function on Z and returns the calculated value for prediction.

2. g() Method: This function takes two variables, X the input data matrix and weights w. It simply performs the dot product of the input matrix and weights and calls the softmax function passing this computation. 

3. Train Method: The train function takes 4 variables, X the input data matrix, T the output variables, the number of iterations and the learning rate. Following steps are performed:
    1. The input data matrix is appended with a column of 1s in the beginning. The weights variable w is initialized with random numbers between 0 and 1. 
    2. A for loop with maximum iterations is initialized. Inside the loop, we find the predicted values by calling the g() function.
    3. Next we calculate update the weights with <b>  w += alpha * X1.T @ (T - ys)</b>. We return the weights at the end of the function. 

4. Use Method: The function takes two variables, X the input data matrix, and the weights. Following steps are performed:
    1. g() method is called to get the complete matrix of prediction.
    2. The np.argmax() function is applied to the above matrix to do actual prediction. This prediction is returned from the function.


### Super Classs Definition

```python
import numpy as np
import abc
from abc import ABC, abstractmethod

# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    

class Classifier(BaseModel):
    """
        Abstract class for classification 
        
        Attributes
        ==========
        meanX       ndarray
                    mean of inputs (from standardization)
        stdX        ndarray
                    standard deviation of inputs (standardization)
    """

    def __init__(self, ):
        self.meanX = None
        self.stdX = None

    def normalize(self, X):
        """ standardize the input X """
        
        if not isinstance(X, np.ndarray):
            X = np.asanyarray(X)

        self.meanX = np.mean(X, 0)
        self.stdX = np.std(X, 0)

        # TODO: Finish this normalization
        Xs = (X-self.meanX)/self.stdX
        return Xs

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abc.abstractmethod
    def train(self, X, T):
        pass
    
    @abc.abstractmethod
    def use(self, X):
        pass 
```

### Pocket Algorithm


```python
class PocketAlgo(Classifier): 
    """ 
        Pocket Algo class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        BaseModel.__init__(self)
        
    def compare(self,X, T, w, wp):
        y = np.sign(X @ w)
        yp = np.sign(X @ wp)
        return 1 if np.sum(y == T) >= np.sum(yp == T) else -1
            
    # train pocket model
    def train(self, X, T,maxiter,alpha):
      w = np.zeros(X.shape[1])
      w_pocket = np.copy(w)
      N = X.shape[0]
     # plt.plot(T)
      for i in range(maxiter):
          converged = True
          for k in np.random.permutation(N): #range(N):
              y = w @ X[k]
              if np.sign(y) != np.sign(T[k]):
                w += alpha * T[k] * X[k]
                converged = False
                ##### additional for pocket algorithm 
              if self.compare(X, T, w, w_pocket) > 0: 
                 w_pocket[:] = w[:]
        
          if converged:
             print("converged at ", i)
             break
      return w_pocket
    # apply the pcoket model to data X
    def use(self, X,w):
        y=X @ w
        prediction=np.array([np.where(y > 0,1,0)])
        return prediction
     
```

## Evaluation Metrics

We need evaluation metrics to calibrate "how good is the model" or in simple terms find the accuracy of a model. These metrics can be used to compare different models or also to identify what can be improved in the model to improve the classification accuracy. We will start with a simple and most intuitive metric. 

$$
   \text{Accuracy} = \frac{\text{The number of correct classification}}{\text{The number of samples}}.
$$

This metric gives us the overall accuracy of the model but does not tell anything about the accuracy in each class. To have a better intuition about the class-wise classification, we will discuss the next metric called confusion matrix.

#### Confusion Matrix
In confusion Matrix, each row represents the predictions from model and column represents the actual classes. The confusion matrix for a binary classification looks like this: 
Here is an example confusion matrix. 

<img src=https://www.dataschool.io/content/images/2015/01/confusion_matrix2.png>


1. <b>TP or True positives</b> are the cases when the actual class of the data point was 1(True) and the predicted is also 1(True)
2. <b>Tn or True negatives</b> are the cases when the actual class of the data point was 0(False) and the predicted is also 0(False)
3. <b>FP or False positives </b> are the cases when the actual class of the data point was 0(False) and the predicted is 1(True)
4. <b>FN or False negatives</b> are the cases when the actual class of the data point was 1(True) and the predicted is 0(False)

Accuracy of a confusion matrix is calculated as: 

<img src=https://cdn-images-1.medium.com/max/1600/1*5XuZ_86Rfce3qyLt7XMlhw.png>

Based on the confusion matrix, we have the following metrics of evaluation:

1. <b>Precision</b>: Precision measures the accuracy of positive predictions.

2. <b>Recall or Sensitivity</b>:  Recall measures how well does the model cover the positive class.

Now, based on Precision and recall we have two more metrics. 

1. <b>F1 Score:</b> Instead of using two metrics, F1 is calculated as the harmonic mean of precision and recall as it provides one value of measure.
$$
\begin{align*}
F_1 &= \Bigg( \frac{\frac{1}{precision} + \frac{1}{recall}}{2} \Bigg)^{-1} \\
\\
    &= 2 \times \frac{precision \times recall}{precision + recall} \\
    \\
    &= \frac{TP}{TP + \frac{FN+FP}{2}}
\end{align*}
$$

<b>If either one is really small between precision and recall, the F1 Score raises a flag and is closer to the smaller value than the bigger one, giving the model an appropriate score rather than just an arithmetic mean.</b>
